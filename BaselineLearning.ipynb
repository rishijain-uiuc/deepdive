```python
from urllib.request import urlopen
import pickle
import numpy

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error as mse
```

```python
df = pickle.load(urlopen('https://drive.usercontent.google.com/u/0/uc?id=1hN_1LAMCoyRu_WXzSj1k9kOmM98zgiBm&export=download'))
df
```

Linear regression  was chosen as the baseline model to run on the data. The features are all the columns in the dataframe, except date and price. The label is the price.

```python
df = df.dropna()
lin_reg = LinearRegression()
columns = list(df.columns)
for r in ['Actual Price', 'Date']:
  columns.remove(r)

X_train, X_test, Y_train, Y_test = train_test_split(df[columns], df['Actual Price'], test_size=0.2)
lin_reg.fit(X_train, Y_train)

Y_pred = lin_reg.predict(X_test)
```

Comparing the predicted price against the actual data, the following mean square error was calculated:

```python
mserr = mse(Y_pred, Y_test)
print("Mean Squared Error: ", mserr)
```
